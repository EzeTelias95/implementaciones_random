```python
#Detect duplicates
from typing import Callable, Iterable

## Optimal Solution O(N)
def has_duplicates_optimal(lst: Iterable[int]) -> bool:
    unique = set()
    for n in lst:
        if n in unique:
            return True
        unique.add(n)
    return False

## The more Pythonic way
def has_duplicates_pythonic(lst: Iterable[int]) -> bool:
    return len(lst) != len(set(lst))


DuplicateStrategy = Callable[[Iterable[int]], bool]

STRATEGIES: dict[str, DuplicateStrategy] = {
    "optimal": has_duplicates_optimal,
    "pythonic": has_duplicates_pythonic,
}

numbers = [1,2,3,4,5,6,4,7,8,9]
print(STRATEGIES["optimal"](numbers))
print(STRATEGIES["pythonic"](numbers))
```
